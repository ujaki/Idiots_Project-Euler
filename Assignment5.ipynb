- [1번 문제] 클래스와 모듈의 공통점과 차이점에 대해 설명하시오.

- 설명 : 
   - 공통점 : 클래스와 모듈의 공통점은 함수나 변수 이름들을 하나의 단위로 묶어서 사용하도록 허용하는 것이다. 프로그램 개발시에 연관있는 코드들을 묶어서 사용하기 때문에 재사용성과 유지보수에 유리한면이 생긴다.
   - 차이점 : 클래스와 모듈은 항목을 액세스하는 방법이 서로 다릅니다. 클래스와 모듈의 주요 차이점은 클래스는 객체로 인스턴스화될 수 있지만 모듈은 객체로 인스턴스화될 수 없다. 모듈의 데이터 복사본은 하나만 존재할 수 있으므로 프로그램의 한 부분에서 모듈의 공용 변수를 변경하면 프로그램의 다른 부분에서도 해당 변수를 읽을 때 동일한 값을 얻는다. 반면 객체 데이터는 인스턴스화된 각 개체마다 별도로 존재합니다. 뿐만 아니라 표준 모듈과 달리 클래스는 인터페이스를 구현할 수 있습니다  또한 클래스와 모듈은 해당 멤버에 대해 다른 범위를 사용합니다.(즉, 이름공간이 다르다) 한 클래스 내에 정의된 멤버는 해당 클래스의 특정 인스턴스 내에서 범위가 지정되며 해당 개체의 수명이 지속되는 동안에만 존재합니다. 클래스 외부에서 클래스 멤버에 액세스하려면 public 형식의 접근 메소드가 존재해야 한다. (단,  멤버가 private로 선언 되어 있는 멤버에 한하여) 그러나 모듈 내에서 선언된 멤버는 기본적으로 공개적으로 액세스할 수 있으며 해당 모듈에 액세스할 수 있는 모든 코드로 액세스할 수 있습니다. 따라서 모듈의 변수는 해당 프로젝트의 어디에서나 볼 수 있으며 프로그램의 수명이 지속되는 동안 존재하므로 전역 변수에 해당합니다.

# Problem 2

- [2번 문제] 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

-다형성이란 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각가 다르게 반응하도록 하는 기능이다.(같은 메시지에 대해 클래스에 따라 다른 행위를 하게 되는 특징이다.) 일반적으로 같은 이름을 가지는 메서드에 대해 인자의 개수와 데이터형에 따라 수행되는 행위가 달라짐을 의미한다. 특히 형 선언이 없다는 점에서  파이썬은 다형성의 장점이 더욱 부각된다. 실시간으로 객체의 형이 결정되므로 단하나의 메소드에 의해 처리될 수 있는 객체의 수는 더욱 많아진다. 

```python
class Animal:
    def Name(self):
        pass
    def Sleep(self):
        print 'sleep'
    def MakeNoise(self):
        pass
    
class Dog(Animal):
    def Name(self):
        print 'I am a dog!'
    def MakeNoise(self):
        print 'Woof!'

class Cat(Animal):
    def Name(self):
        print 'I am a cat!'
    def MakeNoise(self):
        print 'Meow'

class Lion(Animal):
    def Name(self):
        print 'I am a lion'
    def MakeNoise(self):
        print 'Roar'

class TestAnimals:
    def PrintNmae(self, animal):
        animal.Name()
    def GotoSleep(self, animal):
        animal.Sleep()
    def MakeNoise(self, animal):
        animal.MakeNoise()
        
TestAnimals = TestAnimals()
dog = Dog()
cat = Cat()
lion = Lion()

TestAnimals.PrintNmae(dog)
TestAnimals.GotoSleep(dog)
TestAnimals.MakeNoise(dog)
TestAnimals.PrintNmae(cat)
TestAnimals.GotoSleep(cat)
TestAnimals.MakeNoise(cat)
TestAnimals.PrintNmae(lion)
TestAnimals.GotoSleep(lion)
TestAnimals.MakeNoise(lion)
```

설명 : Animal클래스를 정의하고 Dog, Cat, Lion클래스에 상속을 해주었다.
다형성을 이용하여 상속받은 함수 Name, MakeNoise함수를 각 클래스의 호출에 따라서 다르게 기능이 작동하는 것을 볼 수 있습니다.

    > \>\>\> c.incr()  <br/>

  - 요구사항 4. \_\_call\_\_() 메소드를 추가하여 인스턴스 객체를 직접 호출할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때의 동작방식은 incr() 메소드를 호출하였을 때와 동일하다. 
    > \>\>\> c()  <br/>
    > \>\>\> d()  <br/>
    > \>\>\> print c  <br/>
    > [Count (step: 1)] 12 <br/> 
    > \>\>\> print d  <br/>
    > [Count (step: 2)] 14 <br/>
    
  - 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 관련 메소드를 추가하시오.
    > \>\>\> c = c + 5  <br/>
    > \>\>\> d = d - 5  <br/>
    > \>\>\> print c  <br/>
    > [Count (step: 1)] 17 <br/> 
    > \>\>\> print d  <br/>
    > [Count (step: 2)] 9 <br/>  
    
  - 요구사항 6. 다음과 같은 관계연산 (+, -)이 수행될 수 있도록 \_\_cmp\_\_() 메소드를 추가하시오.
    > \>\>\> print c > 10  <br/>
    > True <br/> 
    > \>\>\> print d > 10 <br/>
    > False <br/>
    > \>\>\> print c < 10  <br/>
    > False <br/> 
    > \>\>\> print d < 10 <br/>
    > True <br/>
    > \>\>\> print c == 17  <br/>
    > True <br/> 
    > \>\>\> print d != 9 <br/>
    > False <br/>

```python
# -*- coding: utf-8 -*-
class Counter:
    def __init__(self, count, step = 1): #요구사항 1
        self.count = count
        self.step = step
    def __str__(self): #요구사항 2
        return '[Count (step: ' + str(self.step) + ')] ' + str(self.count)
    def incr(self): #요구사항 3
        self.count = self.count + self.step
    def __call__(self): #요구사항 4
        return self.incr()
    def __add__(self, x): #요구사항 5(+)
        self.count += x
        return Counter(self.count, self.step)
    def __sub__(self, x): #요구사항 5(-)
        self.count -= x
        return Counter(self.count, self.step)
    def __cmp__(self, other): #요구사항 6
        return self.count - other

#-----------------------요구사항 1-----------------------------------------        
print '요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.'
print 'c = Counter(10)'
print 'd = Counter(10, 2)'
c = Counter(10)
d = Counter(10, 2)
print
#-----------------------요구사항 2-----------------------------------------
print '요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 구현하시오.'
print '>>> print c'
print c
print '>>> print d'
print d
print
#-----------------------요구사항 3-----------------------------------------
print '요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 구현하시오'
print '>>> c.incr()'
c.incr()
print '>>> d.incr()'
d.incr()
print '>>> print c'
print c
print '>>> print d'
print d
print
#-----------------------요구사항 4-----------------------------------------
print '요구사항 4. __call__() 메소드를 추가하여 인스턴스 객체를 직접 호출할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때의 동작방식은 incr() 메소드를 호출하였을 때와 동일하다.'
print '>>> c()'
c()
print '>>> d()'
d()
print '>>> print c'
print c
print '>>> print d'
print d
print
#-----------------------요구사항 5-----------------------------------------
print '요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 관련 메소드를 추가하시오.'
print '>>> c = c + 5'
print '>>> d = d - 5'
c = c + 5
d = d - 5
print '>>> print c'
print c
print '>>> print d'
print d
print
#-----------------------요구사항 6-----------------------------------------
print '요구사항 6. 다음과 같은 관계연산 (+, -)이 수행될 수 있도록 __cmp__() 메소드를 추가하시오.'
print '>>> print c > 10'
print c > 10
print '>>> print d > 10'
print d > 10
print '>>> print c < 10'
print c < 10
print '>>> print d < 10'
print d < 10
print '>>> print c == 17'
print c == 17
print '>>> print d != 9'
print d != 9
```

- 설명 : Counter 클래스를 만들고 각 요구사항에 맞는 함수들을 정의하였습니다. 클래스 생성자 함수로서 __init__함수에 인자로 count와 step을 선언하고 문제 조건에 맞게 step을 1로 기본값을 주엇습니다. print로 객체를 출력하면 __str__함수가 호출이 됩니다. 출력 형식에 맞게 해당 객체가 가지고 있는 step과 count값을 출력하는 문장을 리턴해줍니다. incr()함수는 step값 만큼 더하여 다시 count에 저장시킵니다. __call__함수는  인스턴스 객체를 직접 호출할때 호출되는 함수이므로 동작은 incr()함수와 동일하여 call함수 안에서 incr()함수를 호출하여 얻은 값을 반환해줍니다. + 와 -에 대한 함수는 __add__와 __sub__으로 정의되어있어서 입력한 피연산자에 따라서 add 또는 sub 함수가 호출됩니다. 비교연산자가 오면 __cmp__함수가 호출되어서 객체가 가지고 있는 count와 입력한 비교 값과 비교한 결과를 리턴해줍니다.

- [4번 문제] 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init\_\_(), \_\_str()\_\_(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
  - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.
***

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

설명 : 
    -  __name__ == "__main__"이 코드는 현재의 모듈이 최상위 모듈로서 수행 되는지, 아니면 다른 모듈에 의해 import되어 수행되는지르르 구별하기 위해 주로 사용한다.
    - MySet클래스는 list를 상속 받아서 정의를 하고 있다. 클래스 내부에는 생성자 함수인 __init__(), 문자열 표현 함수인  __str__(), 리스트 내의 중복 요소들을 제거해주는 함수 eliminate_duplicate()로 구성되어 있다. 특히 중복 제거 함수는 정적 메소드로 선언이 되어 있어서 인스턴스 객체와 무관하게 일반적으로 클래스 이름을 이용하여 직접 호출할 수 있다.(MySet.eliminate_duplicate(self))
    - __init__() : 리스트 타입의 자료를 인자로 받아서 리스트 내의 요소들을 self에 추가 시킨다. 다 추가가 되면 중복 제거 함수를 불러서 self 리스트 내의 중복 요소들을 제거 한다.
    - __str__() : self 리스트에 저장된 요소들을 출력시켜주는 함수이다.
    - eliminate_duplicate() : 인자로 받은 리스트트 안의 요소들을 공백 리스트인 s에다가 중복이 되지 않는 요소들만 집어 넣고 인자로 받은 리스트는 공백으로 만들어서 중복 제거된 요소드만 다시 l에 저장시킨다 l은 self임으로 중복 제거가 실행된 인스턴스의 self에는 집합의 조건인 중복 없는 요소들만 들어가 있게 된다.

# Problem 5

- [5번 문제] 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오 (정답을 각 요구사항별로 입력할 필요 없이 5번 문제에 대해 1개의 MySet 클래스 정의 코드를 제시하면 된다.)
  - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
    > \>\>\> s = s | t  <br/>
    > \>\>\> print s  <br/>
    > MySet: {1 ,2 ,3 ,4 ,5 ,6 ,7 ,8 ,9} <br/> 
    
  - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> s = s & t  <br/>
    > \>\>\> print s  <br/>
    > MySet: {3} <br/>

  - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> s = s - t  <br/>
    > \>\>\> print s  <br/>
    > MySet: {1, 2} <br/>
***

```python
# -*- coding: utf-8 -*-
class MySet(list):
    def __init__(self, l): #생성자 함수
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self): 
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
        
    def __or__(self, l): #합집합
        res = self[:]
        for x in l:
            if x not in res:
                res.append(x)
        return MySet(res)
        
    def __and__(self, l): #교집합
        res = []
        for x in self:
            if x in l:
                res.append(x)
        return MySet(res)
            
    def __sub__(self, l): #차집합
        for x in self:
            if x in l:
                self.remove(x)
        return MySet(self)
        

    @staticmethod    
    def eliminate_duplicate(l): #중복 원소 제거 함수
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print '>>> s = MySet([1, 2, 2, 3])'
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print '>>> t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])'
    print t
    print
#-----------------------요구사항 1----------------------------------------- 
    print '요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.'   
    print '>>> s = s|t'
    s = s|t
    print '>>> print s'
    print s
    print 
#-----------------------요구사항 2----------------------------------------- 
    print '요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.'
    print '>>> s = MySet([1, 2, 3])'
    print '>>> t = MySet([3, 4, 5])'
    print '>>> s = s & t'
    print '>>> print s'
    s = MySet([1,2,3])
    t = MySet([3,4,5])
    s = s&t
    print s
    print
#-----------------------요구사항 3-----------------------------------------  
    print '요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.'
    print '>>> s = MySet([1, 2, 3])'
    print '>>> t = MySet([3, 4, 5])'
    print '>>> s = s - t'
    print '>>> print s'
    s = MySet([1,2,3])
    t = MySet([3,4,5])
    s = s-t
    print s
    print
#-----------------------문제 6번---------------------------------------- 
    print '문제 6번'
    print '>>> s = MySet([1, 2, 3, 4, 5, 6])'
    s = MySet([1,2,3,4,5,6])
    print '>>> print len(s)'
    print len(s)
    print '>>> print bool(s)'
    print bool(s)
    print '>>> print 2 in s'
    print 2 in s
```

설명  : 피연산자 |, &, -를 사용하면 각각 __or__, __and__, __sub__함수가 호출이 된다.
    - 합집합(__or__) : self에 저장되어 있는 모든 원소들을 res에 저장하고 t와 중복되지 않는 것들을 추가시키면 합집합이 된다.
    - 교집합(__and__) : res를 공백 리스트로 설정하고, s와 t에 동시에 존재하는 원소들을 res에 저장시키면 교집합이 된다.
    - 차집합(__sub__) : s에 존재하는 self리스트의 원소 중에서 t에 있다면 그 원소를 제거시키면 차집합이 된다.
